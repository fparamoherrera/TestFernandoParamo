```python
# El siguiente código crea las tablas necesarias en el esquema MemSch de la base de datos BDDEACEROQA.
import pyodbc

# Conexión a la base de datos BDDEACEROQA
try:
    conn = pyodbc.connect(
        r'DRIVER={ODBC Driver 17 for SQL Server};'
        r'SERVER=Fernando-2020\SQLEXPRESS;'
        r'DATABASE=BDDEACEROQA;'
        r'Trusted_Connection=yes;'
    )
    cursor = conn.cursor()
    print("✅ Conexión exitosa a BDDEACEROQA")
except Exception as e:
    print("❌ Error al conectar:", e)

# Consulta para crear el esquema si no existe
cursor.execute("""
IF NOT EXISTS (SELECT * FROM sys.schemas WHERE name = 'MemSch')
BEGIN
    EXEC('CREATE SCHEMA MemSch')
END
""")

# Crear tabla MemTraMDADet
cursor.execute("""
CREATE TABLE MemSch.MemTraMDADet (
    idMDA INT IDENTITY(1,1) NOT NULL PRIMARY KEY,
    claNodo VARCHAR(10) NOT NULL,
    fecha DATE NOT NULL,
    hora TINYINT NOT NULL,
    pml DECIMAL(10,5) NULL,
    pml_ene DECIMAL(10,5) NULL,
    pml_per DECIMAL(10,5) NULL,
    pml_cng DECIMAL(10,5) NULL,
    FechaUltimaMod DATETIME NULL,
    NombrePcMod NCHAR(30) NULL,
    ClaUsuarioMod INT NULL,
    CONSTRAINT PK_MemTraMDADet PRIMARY KEY (claNodo, fecha, hora)
)
""")

# Crear tabla MemTraMTRDet
cursor.execute("""
CREATE TABLE MemSch.MemTraMTRDet (
    idMTR INT IDENTITY(1,1) NOT NULL PRIMARY KEY,
    claNodo VARCHAR(10) NOT NULL,
    fecha DATE NOT NULL,
    hora TINYINT NOT NULL,
    pml DECIMAL(10,5) NULL,
    pml_ene DECIMAL(10,5) NULL,
    pml_per DECIMAL(10,5) NULL,
    pml_cng DECIMAL(10,5) NULL,
    FechaUltimaMod DATETIME NULL,
    NombrePcMod NCHAR(10) NULL,
    ClaUsuarioMod INT NULL,
    CONSTRAINT PK_MemTraMTRDet PRIMARY KEY (claNodo, fecha, hora)
)
""")

# Crear tabla MemTraTcDet
cursor.execute("""
CREATE TABLE MemSch.MemTraTcDet (
    idTc INT NULL,
    fecha DATE NOT NULL,
    valor DECIMAL(10,6) NULL,
    FechaUltimaMod DATETIME NULL,
    NombrePcMod VARCHAR(30) NULL,
    ClaUsuarioMod INT NULL,
    CONSTRAINT PK_MemTraTcDet PRIMARY KEY (fecha)
)
""")

# Crear tabla MemTraTBFinVw (como tabla física)
cursor.execute("""
CREATE TABLE MemSch.MemTraTBFinVw (
    fecha DATE NULL,
    TbFin NUMERIC(38,14) NULL,
    TbFinTGR NUMERIC(38,9) NULL
)
""")

# Confirmar y cerrar
conn.commit()
cursor.close()
conn.close()

print("Tablas creadas correctamente en SQL Server.")
```
